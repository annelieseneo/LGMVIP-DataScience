```python
import cv2 as cv2 # import OpenCV Python library
from google.colab.patches import cv2_imshow # import patches for image display

DATA_DIR = "/content/gdrive/MyDrive/Ann (5).jpg" # specify directory path
image = cv2.imread(DATA_DIR) # read the image at the directory path

# calculate the ratio of the new image to the old image
r = 300 / image.shape[1] # index 1 indicates width, formula to obtain aspect ratio
dim = (300, int(image.shape[0] * r)) # construct new dimension using same aspect ratio
# prevent the image from being skewed or distorted 

# resize the image according to the specified dimension
image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA) 
cv2_imshow(image) # show image
cv2.waitKey(0) # window will be displayed indefinitely
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert coloured to gray scale
inverted_image = 255 - gray_image # invert to negative image

# blur at Gaussian Kernel Size of odd [height width]
blurred = cv2.GaussianBlur(inverted_image, (11, 11), 0) # at 11 height and width
inverted_blurred = 255 - blurred # convert to inverted blurry image

# mix grayscale image with inverted blurry image to obtain pencil sketch
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0) 
cv2_imshow(pencil_sketch) # show pencil sketch
cv2.waitKey(0) # window will be displayed indefinitely
```
